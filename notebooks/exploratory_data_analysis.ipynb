# Exploratory Data Analysis

This notebook contains exploratory data analysis (EDA) for the customer churn dataset. The goal of this EDA is to understand the dataset, visualize key features, and identify patterns that may help in predicting customer churn.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style='whitegrid')
```

```python
# Load the dataset
data = pd.read_csv('../data/churn_data.csv')

# Display the first few rows of the dataset
data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Visualize the distribution of the target variable (Churn)
plt.figure(figsize=(8, 5))
sns.countplot(x='Churn', data=data)
plt.title('Distribution of Churn')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize correlations between features
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Analyze categorical features
categorical_features = data.select_dtypes(include=['object']).columns
for feature in categorical_features:
    plt.figure(figsize=(10, 5))
    sns.countplot(x=feature, hue='Churn', data=data)
    plt.title(f'Distribution of {feature} by Churn')
    plt.show()
```

## Conclusion

This exploratory data analysis provides insights into the customer churn dataset. Key patterns and correlations have been identified, which will inform the feature selection and model training processes.
